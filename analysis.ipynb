```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('my_data2.csv', encoding='unicode_escape')
```

```python
data.head(10)
```

```python
data['sold'].astype(int)
data= data[data['sold']>0]
data= data[data['sale percentage']>=0]
```

```python
plt.figure(figsize = (10, 5))
data.head()
tmp = data[data['sale_price']>0]
plt.boxplot([tmp['sale_price'],data['price']],labels=['Sale Price', 'Price'])
plt.show()
```

```python
sale_median = tmp['sale_price'].median()
median = tmp['price'].median()

# Q1
sale_q1 = tmp['sale_price'].quantile(0.25)
q1 = tmp['price'].quantile(0.25)

# Q2
sale_q2 = tmp['sale_price'].quantile(0.5)
q2 = tmp['price'].quantile(0.5)

# Q3
sale_q3 = tmp['sale_price'].quantile(0.75)
q3 = tmp['price'].quantile(0.75)

print(f"중앙값: {sale_median}, {median}")
print(f"Q1: {sale_q1}, {q1}")
print(f"Q2: {sale_q2}, {q2}")
print(f"Q3: {sale_q3}, {q3}")
```

```python
data = data.sort_values(by='sold' ,ascending=False)
data.tail(10)
plt.figure(figsize = (10, 5))
plt.scatter(data['sale percentage'],data['sold'])
plt.xlabel('Sale Percentage(%)')
plt.ylabel('Sold')
# plt.title('Scatter p')
plt.show()
```

```python
top_10 = data.head(10)
plt.figure(figsize = (10, 5))
plt.bar(top_10["name"], top_10["sold"])
plt.xlabel("Sale Percentage(%)")
plt.ylabel("Sold")
plt.xticks(top_10['name'],top_10['sale percentage'])
plt.title('Top 10 Sold')
plt.show()
```
